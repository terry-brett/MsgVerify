```python
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
import tensorflow as tf
import numpy as np

#model_name = "typeform/distilbert-base-uncased-mnli"
```

```python
tokenizer = AutoTokenizer.from_pretrained("tokenizer") #loading tokenizer (make sure that tokenizer files are in the folder nameed tokenizer)
```

```python
interpreter = tf.lite.Interpreter(model_path="distilbert_mnli_dynamic_quant.tflite") #loading model
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
premise = "Please send your account details now."
hypothesis = "The text demands personal information"
```

```python
MAX_LEN = 128
inputs = tokenizer(premise, hypothesis, return_tensors="np", padding="max_length", max_length=MAX_LEN, truncation=True)

inputs = {k: v.astype("int32") for k, v in inputs.items()}

interpreter.set_tensor(input_details[0]['index'], inputs["input_ids"])
interpreter.set_tensor(input_details[1]['index'], inputs["attention_mask"])
interpreter.invoke()

output = interpreter.get_tensor(output_details[0]['index'])
print(output)   # logits for [entailment, neutral, contradiction]
```
